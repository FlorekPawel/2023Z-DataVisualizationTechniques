```python
data = json.loads(json_data)
```

```python
G = nx.Graph()
```

```python
for idx, node in enumerate(data["nodes"]):
    G.add_node(idx, name=node["name"], value=node["value"], color=node["colour"])
```

```python
for link in data["links"]:
    G.add_edge(link["source"], link["target"], value=link["value"])
```

```python
node_names = nx.get_node_attributes(G, 'name')
node_values = nx.get_node_attributes(G, 'value')
node_colors = nx.get_node_attributes(G, 'color')
```

```python
edge_values = nx.get_edge_attributes(G, 'value')
```

```python
pos = nx.spring_layout(G)

max_weight = max(edge_values.values())

fig, ax = plt.subplots(1, 1, figsize=(16, 16), dpi=300)

ax.set_title("Star Wars: Episode 1 - The Phantom Menace network", fontsize=20)

node_sizes = [G.nodes[node]['value'] for node in G.nodes]
node_sizes_normalized = np.array(node_sizes) / max(node_sizes)

nx.draw_networkx_nodes(G, pos=pos, node_color=list(node_colors.values()), node_size=node_sizes_normalized * 500, ax=ax)

edge_widths = [G.edges[edge]['value'] for edge in G.edges]
edge_widths_normalized = np.array(edge_widths) / max_weight

nx.draw_networkx_edges(G, pos=pos, width=np.sqrt(edge_widths_normalized) * 10, edge_color="gray", alpha=0.5, ax=ax)

for i, (node, (x, y)) in enumerate(pos.items()):
    ax.text(x, y - 0.02, node_names[i], fontsize=7 + np.log(node_values[i]), ha="center", va="center")

for w in [1, 2, 5, 10, 20]:
    ax.plot([], [], "gray", linewidth=np.sqrt(w / max_weight) * 10, label=w)

ax.legend(title="Number of interactions", frameon=False, bbox_to_anchor=(0, 0.1), loc="lower left")
ax.text(0.01, 0.1,
f"""Colours other than gray represent more important characters in the movie. \nSize of the node depends on the number of interactions in total. \nSize of the link can be rounded.""",
    fontsize=8,
    transform=ax.transAxes,
    verticalalignment="top" )

plt.axis("off")
plt.savefig('output.png', format='png', dpi=300)
plt.show()
```
